```python
# load titanic data set
import pandas as pd
path = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Practice/master/Data/titanic.csv'
titanic = pd.read_csv(path,index_col='PassengerId')
titanic.head(5)
```

Features:
- Survived (0) no, (1) yes
- Pclass: passenger class (first, second, third)
- Name: Passenger's name
- Sex: Male or female
- Age: age
- SibSp: # of siblings/spouses aboard
- Parch: # of parents/children aboard
- Ticket: Ticket number, not used
- Fare: ticket fare
- Cabin: not used
- Embarked: port of embarkation (C=Cherbourg,Q=Queenstown,S=Southampton)


numerical features: Pclass, age, SibSp, Parch, Fare


non-numerical (categorical) features: Sex, Embarked



## Ordinal Encoding


```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
encoder.fit(titanic[['Sex']])
encoder.transform(titanic[['Sex']]) # turns into 0s and 1s for Sex
```

```python
encoder.categories_
```

```python
# ordinal encoding assumes there is an ordering of categories, we can change order
genders = ['male', 'female']
encoder = OrdinalEncoder(categories=[genders])
encoder.fit_transform(titanic[['Sex']])
```

```python
# nan values are set to -1 instead of nan
# unknown values set to 4
encoder = OrdinalEncoder(encoded_missing_value=-1, handle_unknown='use_encoded_value', unknown_value=4) 
encoder.fit_transform(titanic[['Embarked']])
```

```python
# figure out what is 0, 1, 2
encoder.categories_
```

```python
# create a new set with an unknown data entry
new_data = pd.DataFrame(['C', 'C', 'S', 'C', 'Q', 'the Moon'], columns=['Embarked'])
new_data
```

```python
encoder.transform(new_data)
```

# Normalization/standardization (scaling)

* Min-Max Scaling: max - min. Values go from from 0(min) to 1(max)
* Max-abs Scaling: value / (max value). Values go from -1 to 1
* Standard Scaling: (value - mean) / standard deviation

```python
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler
```

```python
titanic[['Age', 'Fare']].head()
```

```python
scaler = MinMaxScaler()
scaler.fit_transform(titanic[['Age', 'Fare']])
```

## Data Imputation


```python
from sklearn.impute import SimpleImputer
```

```python
# missing values can be imputed (replaced) with: a constant value, mean, median, most frequent, etc
imputer = SimpleImputer(strategy='mean')
imputer.fit_transform(titanic[['Age']])
```

## Polynomial features


```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
titanic[['Pclass', 'Age', 'Fare']]
```

```python
poly_features = PolynomialFeatures(degree=2)
poly_features.fit_transform(titanic[['Pclass', 'SibSp', 'Fare']])
# matrix below has columns: [degree 0, pclass, sibSp, Fare, pclass^2, sibsp^2, fare^2, pclass*sibsp, pclass*fare, sibsp*fare]
```
