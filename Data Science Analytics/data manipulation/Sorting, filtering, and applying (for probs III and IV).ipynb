```python
import pandas as pd
```

## Sorting

```python
# Load the imdb_1000 dataset
path = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Practice/master/Data/imdb_1000.csv'
movies = pd.read_csv(path)
movies.head()
```

```python
# sort column title alphabetically
movies.title.sort_values(ascending=False) # set ascending=True or ascending=False, true by default
```

```python
# sort the column duration
movies.duration.sort_values(ascending=False) # if ascending = true, shortest to longest movies in order
```

```python
# sort WHOLE dataframe by star_rating. This gives rest of columns but sorts by given column
movies.sort_values(by='star_rating', ascending=False).head(20)
```

```python
# if there are ties, sort first by star_rating then duration
movies.sort_values(by=['star_rating', 'duration'], ascending=[False, False]) # ascending with an array input also applies
```

## Filtering

```python
# get movies that are at least 200 minutes, then sort by duration
movies.loc[movies.duration >= 200].sort_values(by='duration')
```

```python
# get rows with missing content_rating (shows as NAN in data)
movies.loc[movies.content_rating.isna()]
```

Multiple filter criteria:
- use & instead of and (not &&)
- use | instead of or (not ||)

- add parantheses around each condition

```python
# get all dramas that are atleast 200 minutes long
movies.loc[(movies.genre == 'Drama') & (movies.duration>=200 )]
```

```python
# get crime, drama, and action movies
movies.loc[(movies.genre == 'Drama') | (movies.genre == 'Crime') | (movies.genre == 'Action')]
```

```python
# instead of above which is long, use .isin method. Then filter to ones that are atleast 200 minutes long
movies.loc[(movies.genre.isin(['Crime', 'Drama', 'Action']) & (movies.duration >= 200))]
```

## Applying a function

```python
# load titanic data set, passengers of titanic
path = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Practice/master/Data/titanic.csv'
titanic = pd.read_csv(path)
titanic.head()
```

```python
# map female to 1, and male to 0
my_map={'female':1, 'male':0} # method 1: use dictionary
titanic.Sex.map(my_map)
```

```python
#alternatively use a function
def my_map_fun(sex):
    if sex == 'female':
        return 1
    else:
        return 0
titanic.Sex.map(my_map_fun)
```

```python
# calculate length of strings in the 'name' column
titanic.Name.map(len)
```

```python
# round each element in the #fare column to the next integer
import numpy as np
import math
# titanic.Fare.map(math.ceil)
titanic.Fare.map(np.ceil)
```

```python
# extract the last name of each person
titanic.Name[0].split(',')[0]
```

```python
# create a function that gets last name
def get_last_name(name):
    return name.split(',')[0]
titanic.Name.map(get_last_name)
```

```python
# alternatively, use a lambda function. This is much faster 
titanic.Name.map(lambda name:name.split(',')[0])
```
