```python
import pandas as pd
```

```python
path = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Practice/master/Data/movies.csv'
movies = pd.read_csv(path)
movies # this would print the movies data as a table
```

### Header markdown test

```python
# using python len, returnn number of rows
len(movies)
```

```python
# using pandas .shape returns # of rows and columns
movies.shape
```

```python
# create a dictionary of new and old names. Use this to replace names with spaces to instead use underscores
new_column_names = {'director name': 'director_name', 'movie title': 'movie_title'}
# rename columns
movies.rename(new_column_names, axis=1, inplace=True) # axis is 0 for rows, 1 for columns
# inplace is set to true to modify original data set. False creates a new data set
movies
```

## Removing columns or rows

```python
movies.set_index('movie_title', inplace=True)
```

```python
#remove director name column
movies.drop('director_name', axis=1,)
```

```python
# remove multiple columns (axis 1)
movies.drop(['color', 'language'], axis=1)
```

# .loc and .iloc
#### .loc selects columns or rows by name
### .iloc selects columns or rows by position

```python
#select Avatar, The Avengers, and Toy Story by name. Get all the columns
movies.loc[['Avatar', 'The Avengers', 'Toy Story'],:]
```

```python
# select Spectre through Harry Potter and the Half-Blood Prince, and all columns
movies.loc['Spectre':'Harry Potter and the Half-Blood Prince',:]
```

```python
# select columns color, director name, and movie title
movies.loc[:,['color', 'director_name']]
```

```python
# Select avatar, Toy Story, and columns color through budget
movies.loc[['Avatar', 'Toy Story'], 'color':'budget']
```

```python
# all rows, columns 0, 3, and 5
movies.iloc[:,[0,3,5]]
```

```python
#rows 0 through 5, all columns
movies.iloc[0:6, :] # note: prolly don't need to specific : since all columns selected by default
```

# Handling missing values
#### NaN = not a number. Missing value, 

```python
# True = missing value, False = not missing
movies.isna()
```

```python
# count number of missing values in each column. True (is NaN) = 1, false = 0
movies.isna().sum()
```

```python
# if all values missing in a row, drop that row
movies.dropna(axis=0, how='all') # uses inplace for 3rd argument
```

```python
# if any values missing, drop row
movies.dropna(axis=0, how='any')
```

```python
# if director or country are missing in a row, drop that row
movies.dropna(axis=0, how='any', subset=['director_name', 'country'])
```
