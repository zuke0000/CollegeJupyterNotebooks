# Practice problem VIII

```python
url = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Practice/master/Data/US_Crime_Rates.csv'
crime = pd.read_csv(url)
crime.head()
```

What is the type of the columns?

```python
crime.dtypes
```

Convert the type of the column 'Year' to datetime64.

Hint:

- the function `to_datetime` has a `format` parameter. Set it to `format ='%Y'` 

```python
crime.Year = pd.to_datetime(crime['Year'], format ='%Y')
```

Set the 'Year' column as the index of the dataframe

```python
crime.set_index('Year', inplace=True)
crime.head()
```

Calculate the total number of crimes by decade.

Hints:

- Use the time period `10AS` to resample by decade


```python
crime.resample('10AS').sum()
```

What was the most dangerous decade to live in the US?

Hint:

- pandas `idxmax()` method returns the row label of the maximum value

```python
crime.resample('10AS').sum().Total.idxmax()
```
